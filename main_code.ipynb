```python
from autogluon.tabular import TabularPredictor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder

# Your existing code for data preparation...
train_data = pd.read_csv('/content/Train/train.csv')
test_data = pd.read_csv('/content/Test/test.csv')

# Remove the specified columns
columns_to_remove = ['FORCE_2020_LITHOFACIES_CONFIDENCE']
train_data = train_data.drop(columns=[col for col in columns_to_remove if col in train_data.columns])
test_data = test_data.drop(columns=[col for col in columns_to_remove if col in test_data.columns])

# Your sampling and balancing code...
train_sample = train_data.sample(n=60000)

# Apply your class balancing
mask_65000 = train_sample['FORCE_2020_LITHOFACIES_LITHOLOGY'] == 65000
indices_65000 = train_sample[mask_65000].index
indices_to_drop = indices_65000.to_series().sample(frac=0.2)
train_sample = train_sample.drop(index=indices_to_drop)

mask_30000 = train_sample['FORCE_2020_LITHOFACIES_LITHOLOGY'] == 30000
indices_30000 = train_sample[mask_30000].index
indices_to_drop = indices_30000.to_series().sample(frac=0.1)
train_sample = train_sample.drop(index=indices_to_drop)

mask_65030 = train_sample['FORCE_2020_LITHOFACIES_LITHOLOGY'] == 65030
indices_65030 = train_sample[mask_65030].index
indices_to_drop = indices_65030.to_series().sample(frac=0.1)
train_sample = train_sample.drop(index=indices_to_drop)

# Train your model
hyperparameters = {
    'XT': {
        'n_estimators': 800,
        'max_depth': 200,
        'min_samples_split': 5,
        'min_samples_leaf': 2,
        'criterion': 'gini',
    }
}

predictor = TabularPredictor(label='FORCE_2020_LITHOFACIES_LITHOLOGY').fit(
    train_sample, hyperparameters=hyperparameters, holdout_frac=0.2, time_limit=100000
)

# Make predictions
predictions = predictor.predict(test_data)
y_true = test_data['FORCE_2020_LITHOFACIES_LITHOLOGY']

# Calculate accuracy
accuracy = accuracy_score(y_true, predictions)
print(f"Test Accuracy: {accuracy:.4f}")

# Create confusion matrix
cm = confusion_matrix(y_true, predictions)

# Get unique classes and create labels
unique_classes = sorted(list(set(y_true.unique()) | set(predictions.unique())))
class_labels = [str(cls) for cls in unique_classes]

# Normalize confusion matrix to show percentages
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# Create the confusion matrix plot
plt.figure(figsize=(12, 10))

# Create a custom colormap similar to your image
colors = ['white', '#E6F3FF', '#CCE7FF', '#99D6FF', '#66C5FF', '#3399FF', '#0066CC', '#004080']
n_bins = 256
cmap = sns.blend_palette(colors, n_colors=n_bins, as_cmap=True)

# Plot the heatmap
ax = sns.heatmap(cm_normalized,
                 annot=True,  # Show numbers in cells
                 fmt='.2f',   # Format to 2 decimal places
                 cmap=cmap,   # Use custom colormap
                 xticklabels=class_labels,
                 yticklabels=class_labels,
                 cbar_kws={'label': 'Normalized Frequency'},
                 square=True,  # Make cells square
                 linewidths=0.5,
                 linecolor='gray')

# Customize the plot
plt.title('Confusion Matrix - Lithofacies Classification\n(Normalized by True Class)',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Predicted Labels', fontsize=14, fontweight='bold')
plt.ylabel('True Labels', fontsize=14, fontweight='bold')

# Rotate labels if needed
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()

# Also create a non-normalized version showing raw counts
plt.figure(figsize=(12, 10))

ax2 = sns.heatmap(cm,
                  annot=True,
                  fmt='d',  # Integer format for counts
                  cmap=cmap,
                  xticklabels=class_labels,
                  yticklabels=class_labels,
                  cbar_kws={'label': 'Count'},
                  square=True,
                  linewidths=0.5,
                  linecolor='gray')

plt.title('Confusion Matrix - Lithofacies Classification\n(Raw Counts)',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Predicted Labels', fontsize=14, fontweight='bold')
plt.ylabel('True Labels', fontsize=14, fontweight='bold')

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Print detailed classification report
print("\nDetailed Classification Report:")
print(classification_report(y_true, predictions, target_names=[f"Class_{cls}" for cls in class_labels]))

# Print per-class accuracy
print("\nPer-class Accuracy:")
for i, cls in enumerate(unique_classes):
    class_accuracy = cm[i, i] / cm[i, :].sum() if cm[i, :].sum() > 0 else 0
    print(f"Class {cls}: {class_accuracy:.4f} ({cm[i, i]}/{cm[i, :].sum()})")

# Calculate and print overall metrics
precision = np.diag(cm) / np.sum(cm, axis=0)
recall = np.diag(cm) / np.sum(cm, axis=1)
f1_scores = 2 * (precision * recall) / (precision + recall)

print(f"\nOverall Metrics:")
print(f"Macro-averaged Precision: {np.nanmean(precision):.4f}")
print(f"Macro-averaged Recall: {np.nanmean(recall):.4f}")
print(f"Macro-averaged F1-score: {np.nanmean(f1_scores):.4f}")
```

```python
import copy

def add_feature_noise(X, noise_level_percent, random_state=None):
    """Add Gaussian noise to numerical columns in X."""
    np.random.seed(random_state)
    X_noisy = X.copy()
    # Use only numerical columns
    numerical_columns = X_noisy.select_dtypes(include=[np.number]).columns
    for col in numerical_columns:
        std = X_noisy[col].std()
        noise = np.random.normal(0, noise_level_percent/100 * std, size=X_noisy.shape[0])
        X_noisy[col] += noise
    return X_noisy

def add_label_noise(y, noise_level_percent, random_state=None):
    """Randomly flip a given percent of labels in y."""
    np.random.seed(random_state)
    y_noisy = y.copy()
    n_samples = len(y_noisy)
    n_noisy = int(noise_level_percent/100 * n_samples)
    noisy_indices = np.random.choice(n_samples, size=n_noisy, replace=False)
    unique_labels = np.unique(y)
    for idx in noisy_indices:
        original_label = y_noisy.iloc[idx]
        noisy_choices = [label for label in unique_labels if label != original_label]
        y_noisy.iloc[idx] = np.random.choice(noisy_choices)
    return y_noisy

def noise_sensitivity_test(predictor, test_data, noise_levels=[5,10,15,20,25,30]):
    orig_X = test_data.drop('FORCE_2020_LITHOFACIES_LITHOLOGY', axis=1)
    orig_y = test_data['FORCE_2020_LITHOFACIES_LITHOLOGY']
    results = {'Feature':[], 'Label':[], 'Both':[]}
    for noise_pct in noise_levels:
        # Features only
        X_noisy = add_feature_noise(orig_X, noise_pct, random_state=42)
        clf_pred = predictor.predict(X_noisy)
        acc_feature = accuracy_score(orig_y, clf_pred)
        results['Feature'].append(acc_feature)

        # Labels only
        y_noisy = add_label_noise(orig_y, noise_pct, random_state=42)
        acc_label = accuracy_score(y_noisy, predictor.predict(orig_X))
        results['Label'].append(acc_label)

        # Both features & labels
        X_noisy2 = add_feature_noise(orig_X, noise_pct, random_state=42)
        y_noisy2 = add_label_noise(orig_y, noise_pct, random_state=42)
        acc_both = accuracy_score(y_noisy2, predictor.predict(X_noisy2))
        results['Both'].append(acc_both)

        print(f"Noise {noise_pct:2d}% - Feature Noise accuracy: {acc_feature:.4f}, Label Noise accuracy: {acc_label:.4f}, Both accuracy: {acc_both:.4f}")
    return results

# Run the noise sensitivity test!
noise_results = noise_sensitivity_test(predictor, test_data)

# Optionally, plot the results
plt.figure(figsize=(8,5))
for label, accs in noise_results.items():
    plt.plot([5,10,15,20,25,30], accs, marker='o', label=label+' Noise')
plt.xlabel('Noise Level (%)', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Noise Sensitivity Analysis', fontsize=16, fontweight='bold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# === Baseline Comparison with Default Hyperparameters, Encoding + NaN Imputation ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder

# --- Step 1: Load and pre-process your data ---
train_data = pd.read_csv('/content/Train/train.csv')
test_data = pd.read_csv('/content/Test/test.csv')

columns_to_remove = ['FORCE_2020_LITHOFACIES_CONFIDENCE']
train_data = train_data.drop(columns=[c for c in columns_to_remove if c in train_data.columns])
test_data = test_data.drop(columns=[c for c in columns_to_remove if c in test_data.columns])

# Sampling
train_sample = train_data.sample(n=60000)

# Balancing
mask_65000 = train_sample['FORCE_2020_LITHOFACIES_LITHOLOGY'] == 65000
train_sample = train_sample.drop(mask_65000.sample(frac=0.2).index)

mask_30000 = train_sample['FORCE_2020_LITHOFACIES_LITHOLOGY'] == 30000
train_sample = train_sample.drop(mask_30000.sample(frac=0.1).index)

mask_65030 = train_sample['FORCE_2020_LITHOFACIES_LITHOLOGY'] == 65030
train_sample = train_sample.drop(mask_65030.sample(frac=0.1).index)

# --- Step 2: Split features/labels ---
X_train = train_sample.drop(columns=['FORCE_2020_LITHOFACIES_LITHOLOGY'])
y_train = train_sample['FORCE_2020_LITHOFACIES_LITHOLOGY']

X_test = test_data.drop(columns=['FORCE_2020_LITHOFACIES_LITHOLOGY'])
y_test = test_data['FORCE_2020_LITHOFACIES_LITHOLOGY']

# --- Step 3: Impute missing values ---
def impute_missing(train_df, test_df):
    train_imp = train_df.copy()
    test_imp = test_df.copy()
    for col in train_imp.columns:
        if train_imp[col].dtype == 'object':
            train_imp[col] = train_imp[col].fillna("missing")
            test_imp[col] = test_imp[col].fillna("missing")
        else:
            median_val = train_imp[col].median()
            train_imp[col] = train_imp[col].fillna(median_val)
            test_imp[col] = test_imp[col].fillna(median_val)
    return train_imp, test_imp

X_train_imp, X_test_imp = impute_missing(X_train, X_test)

# --- Step 4: Encode categorical features ---
def encode_categoricals(train_df, test_df):
    train_enc = train_df.copy()
    test_enc = test_df.copy()
    for col in train_enc.columns:
        if train_enc[col].dtype == 'object':
            le = LabelEncoder()
            combined = pd.concat([train_enc[col], test_enc[col]], axis=0).astype(str)
            le.fit(combined)
            train_enc[col] = le.transform(train_enc[col].astype(str))
            test_enc[col] = le.transform(test_enc[col].astype(str))
    return train_enc, test_enc

X_train_enc, X_test_enc = encode_categoricals(X_train_imp, X_test_imp)

# --- Step 5: Define models (defaults) ---
models = {
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(n_jobs=-1),
    "LogisticRegression": LogisticRegression(max_iter=1000, n_jobs=-1),
    "SVM": SVC(),
    "KNeighbors": KNeighborsClassifier(n_jobs=-1)
}

# --- Step 6: Train and Evaluate ---
results = {}
for name, model in models.items():
    print(f"\n=== Training {name} ===")
    model.fit(X_train_enc, y_train)
    preds = model.predict(X_test_enc)

    acc = accuracy_score(y_test, preds)
    macro_prec = precision_score(y_test, preds, average='macro', zero_division=0)
    macro_rec = recall_score(y_test, preds, average='macro', zero_division=0)
    macro_f1 = f1_score(y_test, preds, average='macro', zero_division=0)

    cm = confusion_matrix(y_test, preds)
    results[name] = {
        "accuracy": acc,
        "macro_precision": macro_prec,
        "macro_recall": macro_rec,
        "macro_f1": macro_f1,
        "confusion_matrix": cm
    }

    print(f"{name} Accuracy: {acc:.4f}")
    print(f"Macro Precision: {macro_prec:.4f}")
    print(f"Macro Recall:    {macro_rec:.4f}")
    print(f"Macro F1-Score:  {macro_f1:.4f}")
    print(classification_report(y_test, preds, zero_division=0))

# --- Step 7: Accuracy summary ---
print("\n=== Model Metric Comparison ===")
for name, res in results.items():
    print(f"{name}: Accuracy={res['accuracy']:.4f}, "
          f"MacroPrecision={res['macro_precision']:.4f}, "
          f"MacroRecall={res['macro_recall']:.4f}, "
          f"MacroF1={res['macro_f1']:.4f}")

# --- Step 8: Confusion Matrix Plots ---
unique_classes = sorted(y_test.unique())
for name, res in results.items():
    cm = res["confusion_matrix"]
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
                xticklabels=unique_classes,
                yticklabels=unique_classes)
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()
```
